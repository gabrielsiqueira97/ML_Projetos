# **Base Credit Data**

```python
with open('/content/credit.pkl','rb') as f:
  x_credit_treinamento, y_credit_treinamento, x_credit_teste, y_credit_teste = pickle.load(f)
```

```python
x_credit_teste.shape, y_credit_teste.shape
```

```python
arvore_credit = DecisionTreeClassifier(criterion='entropy', random_state = 0)
arvore_credit.fit(x_credit_treinamento, y_credit_treinamento)
```

```python
previsoes = arvore_credit.predict(x_credit_teste)
previsoes
```

```python
accuracy_score(y_credit_teste, previsoes)
```

```python
cm = ConfusionMatrix(arvore_credit)
cm.fit(x_credit_treinamento, y_credit_treinamento)
cm.score(x_credit_teste, y_credit_teste)
```

```python
fig, ax = plt.subplots(figsize=(25,20))
tree.plot_tree(arvore_credit, feature_names=previsores, filled=True, class_names=['0', '1'])
class_names = [str(c) for c in arvore_credit.classes_]
tree.plot_tree(arvore_credit, feature_names=previsores, filled=True, class_names=class_names)
fig.savefig('arvore_credit.png')
```
